```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pd.pandas.set_option('display.max_columns',None)
df = pd.read_csv('pokemon.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

# Missing Values

```python
plt.figure(figsize=(16,7))
plt.title('Null Values')
sns.heatmap(df.isnull(),yticklabels=False)
```

```python
missing = [feat for feat in df.columns  if df[feat].isnull().sum()>1]

for feat in missing:
    print(f'{feat} -> {np.round(df[feat].isnull().mean(),4)} ')
```

```python
# feature containing Nan

for feat in missing:
    data = df.copy()
    data[feat] = np.where(data[feat].isnull(),1,0)
    
    plt.title(feat)
    sns.countplot(data[feat])
    plt.show()
```

# Numeric values 

```python
numeric = [feat for feat in df.columns if df[feat].dtype != 'O']
```

```python
df[numeric].head()
```

# Discrete Variable

```python
discrete_feat = [feat for feat in numeric if len(df[feat].unique())<25 ]
```

```python
for feat in discrete_feat:
    sns.countplot(df[feat])
    plt.title(feat)
    plt.show()
```

# Continuous Variable

```python
continuous_feat = [feat for feat in numeric if feat not in discrete_feat]
```

```python
for feat in continuous_feat:
    plt.hist(df[feat])
    plt.title(feat)
    plt.show()
```

# categories Variable

```python
categories = [feat for feat in df.columns if df[feat].dtype == 'O']
```

```python
categories
```

```python
df[categories].head()
```

```python
for feat in categories:
    print(f'{feat} ->{len(df[feat].unique())}')
```
